```python
nfcount = 0
notfraud = df['Class']
for i in range(len(notfraud)):
    if notfraud[i] == 0:
        nfcount = nfcount + 1
nfcount
percentage_not_fraud = (nfcount/len(notfraud)*100)
print('Percentage of total not Fraud transaction in the dataset : ',percentage_not_fraud)
```

```python
fcount = 0
fraud_count = df['Class']
for i in range(len(fraud_count)):
    if notfraud[i] == 1:
        fcount = fcount + 1
fcount
percentage_fraud = (fcount/len(fraud_count)*100)
print('Percentage of total Fraud transaction in the dataset : ',percentage_fraud)
```

```python
plot_data=pd.DataFrame()
plot_data['Fraud Transaction'] = fraud_count
plot_data['Genuine Transaction'] = notfraud
plot_data
```

```python
plt.title('bar plot Fraud Transaction and Genuine Transaction',fontsize =8)
sns.barplot(x = 'Fraud Transaction',y ='Genuine Transaction',data = plot_data)
plt.show()
```

```python
x_train,x_test,y_train,y_testt = train_test_split(x, y, test_size = 0.2, random_state = 42) 
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test)
accuracy= log_reg.score(x_test,y_testt)
```
